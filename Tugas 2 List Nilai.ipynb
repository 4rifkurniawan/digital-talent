```python
import random

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print(nilaiTugas)

# TAMPILKAN:
# ganjil
# nilai > 70
# standar deviasi
# maksimum
# minimum
# modus
```

```python
print('---Menampilkan Angka Ganjil dari ke 250 nilai tersebut---')
for n in nilaiTugas :
    if n%2==1 :
        print (n)        
    
```

```python
print ('---banyak yang mendapatkan nilai lebih dari 70---')              
#membuat list baru
list70 = [] 
for m in nilaiTugas :
    if m>70 :
        list70.append(m)
print(list70)
#menghitung jumlah index list70
print('\njumlah nilai > 70 adalah ',len(list70))
```

```python
import math

print('---Mencari Nilai Standar Deviasi---')
print('\nA. list nilaiTugas')
#print(nilaiTugas)

jumlah_nT = 0
for nT in nilaiTugas :
    jumlah_nT=jumlah_nT+nT
print('Jumlah nilai pada List nilaiTugas = ',jumlah_nT)
print('(Jumlah nilai pada List nilaiTugas)^2 = ',jumlah_nT**2)

#----------------------------------------------
print('\nB. membuat list baru list_kuadrat')
list_kuadarat = []
for m in nilaiTugas :
    p = m**2
    list_kuadarat.append(p)
#print(list_kuadarat)

jumlah_lk = 0
for lk in list_kuadarat :
    jumlah_lk=jumlah_lk+lk
print('Jumlah nilai pada List_kuadarat = ',jumlah_lk)
    
print('\nC. Jumlah Data')
n = len(nilaiTugas)
print('n = ',n)
q = n**2-n                   # q = n(n-1)
print(q)

print('\nD. Nilai Standar Deviasi')
s = ((n*jumlah_lk) - (jumlah_nT**2)) / (q)   #nilai varian s
print('Standar Deviasi = ',math.sqrt(s))
```

```python
print('\n---Mencari Nilai Maksimum---')
#print(nilaiTugas)
m=0
for n in nilaiTugas :
    if n >= m :
        m = n
print('nilai maksimum = ',m)

print('\n---Mencari Nilai Minimum---')
#print(nilaiTugas)
p=100
for q in nilaiTugas :
    if  q <= p :
        p = q
print('nilai minimum = ',p)
```

```python
print('---Mencari Modus Nilai Tugas---')
#print(nilaiTugas)

nilaiKeys = set(nilaiTugas)
# print(dictionaryNilaiTugasKeys)
listKeys = list(nilaiKeys)
# print(listKeys)

maksFreq = 0
modusData = []
dictionaryNilai = {}
for k in listKeys:
    n_freq = nilaiTugas.count(k)
    dictionaryNilai[k] = n_freq
    if n_freq>maksFreq:
        maksFreq = n_freq

# print(dictionaryNilai)

for k,v in dictionaryNilai.items():
    if v == maksFreq:
        modusData.append(k)
        
print('Modus data dengan jumlah kemunculan terbanyak yaitu sebanyak =',maksFreq, '\nModus nya adalah sebagai berikut:',modusData)
            
```
