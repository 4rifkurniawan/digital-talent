Pada tahun 2007 - 2013 D.I. Yogyakarta mendapatkan Dana Bantuan Langsung Masyarakat (BLM) dan Kemiskinan dari pemerintah melalui program Program Nasional Pemberdayaan Masyarakat (PNPM) Mandiri Perdesaan. Bantuan Langsung Masyarakat dan Kemiskinan tersebut disalurkan ke seluruh kabupaten dan kota yang ada di wilayah D.I Yogyakarta hingga tingkat kecamatan. Total ada 64 kecamatan yang mendapatkan bantuan langsung masyarakat dari pemerintah antara lain, 8 kecamatan di Kab. Kulon Progo, 17 Kecamatan di Kab. Bantul, 13 Kecamatan di kab. Gunungkidul, 17 Kecamatan di Kab. Sleman dan 14 Kecamatan di Kota Yogyakarta. Bantuan Langsung Masyarakat dan Kemiskinan ini disalurkan secara bertahap mulai dari tahun 2007 hingga 2013. Proses penyaluran bantuan tersebut juga bervariasi. Dalam rentang tahun 2007 hingga 2013 ada kecamatan yang hanya mendapatkan sekali saja dan ada pula kecamatan yang mendapatkan bantuan setiap tahunnya. 
Masyarakat penerima Bantuan langsung tersebutberupa masyarakat yang tergolong Penduduk miskin berdasarkan data Badan Pusat Statistik tentang Pendataan Program Perlindungan Sosial (PPLS) tahun 2011. Pada data Bantuan Langsung Masyarakat dan Kemiskinan di D.I. Yogyakarta juga akan menampilkan kategori dari kecamatan penerima bantuan. Kategori dari kecamatan tersebut terbagi menjadi 3 buah antara lain kategori kecamatan Miskin, Sedang dan Tidak Miskin.  
Terdapat 2 buah sumber dana Bantuan Langsung Masyarakat dan Kemiskinan untuk D.I. Yogyakarta antara lain Anggaran Pendapatan dan Belanja Negara (APBN) dan Anggaran Pendapatan dan Belanja Daerah (APBD). Besarnya sumber pendanaan bervariasi di masing-masing wilayah kecamatan, tetapi secara umum sumber APBN lebih besar dari APBD  

```python
import pandas as pd
data_blm_kemiskinan = pd.read_csv('processed-data-blm-kemiskinan-di-yogyakarta-2007-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_blm_kemiskinan
```

## Analisis Bantuan Langsung Masyarakat per Kabupaten/Kota

```python
data_blm_kemiskinan.groupby(['nama_kabkota'])['jumlah_penduduk_miskin','total_blm'].sum()
```

```python
ax = data_blm_kemiskinan.groupby('nama_kabkota').sum()['total_blm']
ax = ax.plot(kind='bar', title ="Jumlah BLM Berdasarkan Kab/Kota", legend=True, fontsize=10)
ax.set_xlabel("nama_kabkota", fontsize=10)
ax.set_ylabel("total_blm", fontsize=10)
```

```python
ax = data_blm_kemiskinan.groupby('nama_kabkota').sum()['jumlah_penduduk_miskin']
ax = ax.plot(kind='pie', title ="Jumlah Penduduk Miskin Berdasarkan Kab/Kota", legend=True, fontsize=10)
ax.set_xlabel("nama_kabkota", fontsize=10)
ax.set_ylabel("jumlah_penduduk_miskin", fontsize=10)
```

Dalam periode tahun 2007 - 2013, kabupaten Bantul mendapatkan Bantuan Langsung Masyarakat yang lebih tinggi dibandingkan dengan kabupaten dan kota yang ian di D.I. Yogyakarta. Besarnya bantuan yang diterima oleh kabupaten Bantul sebesar 85.785. Hal ini sebanding dengan jumlah penduduk miskin di Kabupaten Bantul yang menerima bantuan langsung masyarakat sebanyak 723.069 jiwa.
Sedangkan kabupaten yang mendapatkan bantuan langsung masyarakat paling kecil adalah Kab. Gunungkidul sebesar 23.650 dan jumlah pendukuk miskin yang menerima bantuan sebnyak 152.453 jiwa. Data total_blm yang disajikan dalam skala juta rupiah.

## Analisis Bantuan Langsung Masyarakat per Tahun

```python
data_blm_kemiskinan.groupby(['tahun'])['jumlah_penduduk_miskin','total_blm'].sum()
```

```python
ax = data_blm_kemiskinan.groupby('tahun').sum()['total_blm']
ax = ax.plot(kind='pie', title ="Jumlah BLM Berdasarkan Tahun Anggaran", legend=True, fontsize=10)
ax.set_xlabel("tahun", fontsize=10)
ax.set_ylabel("total_blm", fontsize=10)
```

Bantuan Langsung Mandiri yang diberikan pemerintah untuk wilayah D.I. Yogyakarta mempunyai nilai bantuan paling besar pada tahun 2009 yaitu sebesar Rp 58.285.000.000,-. Hal ini sebanding dengan jumlah penduduk miskin yang menerima bantuan tersebut yaitu sebanyak 334.145 jiwa. Sedangkan pada tahun 2007, pemerintah memberikan bantuan langsung kepada masyarakat paling sedikit, yaitu sebanyak Rp 4.850.000.000 untuk 27.721 jiwa. 

## Analisis Bantuan Langsung Masyarakat berdasarkan kabupaten per tahun

```python
a = data_blm_kemiskinan.groupby(['nama_kabkota','tahun'])['total_blm'].sum()
a
```

```python
ax = data_blm_kemiskinan.groupby(['nama_kabkota','tahun']).sum()['total_blm']
ax = ax.plot(kind='bar', title ="Jumlah BLM Berdasarkan Tahun per Kab/Kota", legend=True, fontsize=10)

ax.set_xlabel("tahun tiap Kab/kota", fontsize=13)
ax.set_ylabel("total_blm", fontsize=10)
```

Dalam rentang waktu 2007 - 2013 berasarkan analisis berikut Kabupaten Bantul pada tahun 2009 mendapatkan bantuan langsung masyarakat yang paling banyak, yaitu sebesar Rp 22.075.000.000 tetapi tahun-tahun berikutnya semakin menurun bantuan yang diterima. Sedangkan bantuan langsung masyarakat yang paling sedikit diterima oleh kabupaten Kulon Progo pada tahun 2007 yaitu sebesar Rp 900.000.000

## Analisis Bantuan Langsung Masyarakat berdasarkan Kecamatan 

```python
data_blm_kemiskinan.groupby(['nama_kabkota','kecamatan'])['jumlah_penduduk_miskin','total_blm'].sum()
```

```python
ax = data_blm_kemiskinan.groupby(['nama_kabkota','kecamatan']).sum()['total_blm']
ax = ax.plot(kind='line', title ="Jumlah BLM Berdasarkan Tahun per Kab/Kota", legend=True, fontsize=10)

ax.set_xlabel("kecamatan tiap Kab/kota", fontsize=13)
ax.set_ylabel("total_blm", fontsize=10)
```

Dalam periode tahun 2007 - 2013 kecamatan yang paling banyak mendapatkan bantuan langsung masyarakat adalah kecamatan Banguntapan - Kabupaten Bantul sebanyak Rp 9.470.000 dengan jumlah penduduk miskin yang mendapatkan bantuan sebanyak 59.250 jiwa. Sedangkan kecamatan Pakualaman kota Yogyakarta paling sedikit menerima bantuan langsung masyarakat sebanyak Rp. 850.000.000 dengan jumnlah penduduk miskin yang mendapatkan bantuan sebanyak 3.380 jiwa
