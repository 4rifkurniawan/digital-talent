```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

1. Prepare a list of offices sorted by country, state, city. 

```python
run("SELECT * FROM offices ORDER BY country,state,city")
```

2. How many employees are there in the company?

```python
run("SELECT officeCode,COUNT(*) AS Jumlahpekerja FROM employees GROUP BY officeCode")

```

3. What is the total of payments received? 

```python
run("""
SELECT 
    SUM(amount) AS Total_Pembayaran
FROM payments
""")
```

4. List the product lines that contain 'Cars'. 

```python
run("""
SELECT 
    *
FROM
    products
WHERE
    productLine LIKE '%CARS%'
""")
```

5. Report total payments for October 28, 2004.

```python
run("""
SELECT 
    SUM(amount) AS TotalBayar 
FROM payments
WHERE paymentDate='2004-10-28'

""")
```

6. Report those payments greater than $100,000 dolar. 

```python
run("""
SELECT 
    paymentDate,
    amount
FROM payments
WHERE amount>'100000'
ORDER BY paymentDate
""")
```

7. List the products in each product line

```python
run("""
SELECT 
    productName,
    productLine
FROM products
ORDER BY productLine ASC
""")
```

8. How many products in each product line? 

```python
run("""
SELECT 
    productLine,
    COUNT(*) AS Jumlah_produk
FROM products
GROUP BY productLine
ORDER BY productLine
""")
```

9. What is the minimum payment received? 

```python
run("""
SELECT 
    MIN(amount) AS Pembayaran_terendah
FROM payments
""")
```

10. List all payments greater than twice the average payment.

```python
run("""
SELECT 
    paymentDate,
    amount
FROM payments
WHERE amount > 2 * (select AVG(amount) from payments)
ORDER BY paymentDate
""")
```

11. What is the average percentage markup of the MSRP on buyPrice? 

```python
run("""
SELECT 
        AVG(((MSRP-buyPrice)/buyPrice)*100) as Rata2_Presentase_Markup
FROM products
""")
```

12. How many distinct products does ClassicModels sell?

```python
run("""
SELECT 
    COUNT(DISTINCT productCode) AS Jumlah_Distinct_Products
FROM orderdetails 
""")
```

13. Report the name and city of customers who don't have sales representatives? 

```python
run("""
SELECT customers.customerName,customers.city 
FROM customers
INNER JOIN offices 
ON customers.city!=offices.city
ORDER BY city
""")
```

14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting

```python
run("""
SELECT 
    CONCAT (firstName," ",lastName) AS Nama,
    jobTitle
FROM employees
WHERE jobTitle LIKE '%Manager%' OR jobTitle LIKE '%VP%' 
""")
```

15. Which orders have a value greater than 5,000 dolar? 

```python
run("""
SELECT 
    orderNumber,
    SUM(quantityOrdered*priceEach) AS jumlah_harga
FROM orderdetails
WHERE (quantityOrdered*priceEach)>5000
GROUP BY orderNumber
ORDER BY jumlah_harga
""")
```
